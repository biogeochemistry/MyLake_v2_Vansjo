```python
t = np.linspace(0,20,21)
```

```python
def Pfun(P0, t, Fext, Fsed):
    tau = 0.85
    Fx=(Fext+Fsed)*3*tau
    return Fx - (Fx - P0)*np.exp(-t/3/tau)
```

```python
F = 0.0585
plt.plot(t, Pfun(0.3, t, F, F))
plt.plot(t, Pfun(0.3, t, F/10*9, F))
plt.plot(t, Pfun(0.3, t, F/10*8, F))
plt.plot(t, Pfun(0.3, t, F/10*7, F))
plt.plot(t, Pfun(0.3, t, F/10*6, F))
plt.xlim(0,10)
```
